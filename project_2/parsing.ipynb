```python
def parsechunks(chunks):
    chunks = chunks.strip()
    sep_chunk = chunks.split("\n")
    
    username = sep_chunk[0]
    post = int(sep_chunk[1].split("post")[0].replace(",",""))
    followers = float(sep_chunk[2].split("followers")[0].replace(",","").replace("K","").replace("M",""))
    if("K" in sep_chunk[2]):
        folllowers = int(followers * 1000)
    elif("M" in sep_chunk[2]):
        followers = int(followers * 1000000)
    else:
        followers = int(followers)
        
    following = float(sep_chunk[3].split("following")[0].replace(",","").replace("K","").replace("M",""))
    if("K" in sep_chunk[2]):
        folllowing = int(followers * 1000)
    elif("M" in sep_chunk[2]):
        following = int(followers * 1000000)
    else:
        following = int(followers)
    name = sep_chunk[4]
    if(len(sep_chunk) >4):
        types_of_page = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else:
        type_of_pages ="UNKNOWN"
        bio = " "
    return{" username" : username,"post":post,"followers":followers,"following":following,"name":name,"types_of_page":types_of_page,"bio":bio}
#display = parsechunks(chunks[0])
#print(display,sep="\n")
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunks = parsechunks(chunk)
    all_chunks.append(parsed_chunks)
print(all_chunks)
```

```python
import json 

s=json.dumps(all_chunks, indent = 4)
with open("data.json","w") as f:
    f.write(s)
#print(s)
```
